```python
def add(a,b): # a & b called formal argument
    c = a+b
    print(c)
    
add(5,6) #5 and 6 we called as actual argument
```

```python
def count(lst):
    
    even = 0
    odd = 0
    
    for i in lst:
        if i%2 == 0:
            even += 1
        else:
            odd +=1 
    return even,odd

lst = [1, 2, 3, 4, 5 ]
even, odd = count(lst)         

print(even)
print(odd)
```

```python
# in progammin we need to continue these process thats why we need to use loop hear

def fib(n):
    a = 0
    b = 1
    
    print(a)
    print(b)
    
    for i in range(0, n):
        c = a + b
        a = b
        b = c
        
        print(c)
        
fib(10)
```

```python
# factorial number

def fact(n):
    f = 1
    for i in range(1, n+1):
        f = f*i
        
    return f

x = 4
result = fact(x)
print(result)     

# please use debug the code in pycharm for more indetail explanation & breakthrouth point at f = 1
```
